## 4) Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
np.random.seed(192)

mean = 30
sigma = 10
number_of_entries = 12345
data = np.random.normal(mean, sigma, number_of_entries)
plt.hist(data, bins=100, range=(0,120), color="blue")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

### Logarithmic Scale
A logarithmic scale can be used to highlight differences in bins in histograms or points in on graph. This works particular well with skewed graphs such as exponential decay to extract the most out of the beginning data points. 

b)

```python
data_x = data
data_y = np.random.normal(mean, sigma, number_of_entries)
plt.hist2d(data_x, data_y, 100)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

### Color scale
This color scale indicates the z value on the xy plane. The shorter the wavelength of color the higher the z value.

```python
data = np.round(data,0)
bins = int (1 + 3.222 * math.log10(data.size))
print ("Bins calculated with Sturge's Rule = {}".format(bins))
print ("Bins calculated with trial and error {}".format(30))
plt.hist(data, bins=30, range=(0,120), color="blue")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show
```

### Aliasing

The aliasing artifact is appearant in the large spikes of events in bins from the discreet numbers. I decreased the bins from 100 to 30. A few methods can be utilized to eliminate aliasing. While I found simple trial and error to optimize the binning other rules such the square root rule, Sturge's rule, Doane's rule, or other rules attempt to find ideal binning. Some such as Sturge's rule are critisized for over smoothing. This can be seen by the lower number of bins compared to my trial and error method.
